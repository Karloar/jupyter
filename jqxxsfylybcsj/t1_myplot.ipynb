```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 200)
y = np.sin(x)
yn = y + np.random.rand(1, len(y)) * 1.5
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, yn, c='blue', marker='o')
```

```python
ax.plot(x, y+0.75, 'r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist = np.array([[0.1, 0.1], [0.9, 0.5], [0.9, 0.1], [0.45, 0.9], [0.9, 0.8], [0.7, 0.9], [0.1, 0.45], [0.45, 0.1]])
m, n = dist.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(dist[:, 0], dist[:, 1], c='blue', marker='o')
for point in dist.tolist():
    plt.annotate("(" + str(point[0]) + ", " + str(point[1]) + ")", xy=(point[0], point[1]))
ax.plot(dist[:, 0], dist[:, 1], 'r')
plt.show()
```
